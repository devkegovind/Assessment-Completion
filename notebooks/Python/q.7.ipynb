**IMPORT LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**IMPORT DATASET**

```python
df = pd.read_csv("D:\\Downloads\\NASA.csv")
```

```python
df.head()
```

**Get all the Earth meteorites that fell before the year 2000**

```python
df['year'] = pd.to_datetime(df['year'], errors = 'coerce')
```

*Filter the dataset to include only Earth meteorites that fell before the year 2000*

```python
filtered_df = df[(df['nametype'] == 'Valid') & (df['year'].dt.year < 2000)]
```

```python
filtered_df
```

```python
# Univariate Plot - Year Distribution
filtered_df['year'].dt.year.plot(kind='hist', bins=30)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Meteorite Falls by Year')
plt.show()
```

```python
plt.figure(figsize=(35,25))
filtered_df['year'].dt.year.plot(kind='bar', rot=90)
#plt.xticks(rotation = 45)
plt.show()
```

```python
# Bivariate Plot - Latitude vs Longitude
plt.scatter(filtered_df['reclong'], filtered_df['reclat'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Meteorite Falls: Latitude vs Longitude')
plt.show()
```

**Get all the earth meteorites co-ordinates who fell before the year 1970**
